# Zeros of a function - comparision of methods

2023-04-05

This Notebook compares the execution times obtained and also the number of iterations for each implemented method. Each method was implemented separately in a corresponding Notebook: bisection, raphson, and secant.

- **hands-on-01-bisection.ipynb** : the bisection method reduces the interval by half, until it becomes small enough. It is not the fastest method to calculate roots, but it is the most reliable, since bisection will always find the root from the given interval.
- **hands-on-01-secant.ipynb** : in the secant method, the function f(x) is assumed to be approximately linear near the root, and then the next estimated value to find the root can be calculated by linear interpolation.
- **hands-on-01-raphson.ipynb** : the Newton-Raphson method is simple and fast, but has the disadvantage of needing the derivative ùëì(ùë•) of the function. This implementation uses the [SymPy](https://www.sympy.org/en/index.html) library to solve the derivatives of functions, and also to find the result of the function, given the value of the variable.

The implementations use the standard Python environment and Jupyter Notebook and have not been optimized. The codes ran on a machine with an Intel Core i7-9750H processor with a maximum of 4500 MHz.

For future work it would be interesting to optimize implementations for performance, and compare speedus and efficiencies.

        
Notes
